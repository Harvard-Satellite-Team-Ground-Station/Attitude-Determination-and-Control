# Initial Sun Sensor Calculations

Initial configurations of our sun sensors will consist of simple photodiode arrays mounted onto the outward facing sides of our 6U cube satellite. In order to maximize utility while minimizing cost, we will use many coarse sun sensors that are easy to manufacture, mounting a sun sensor on every face of our cube satellite to ensure full coverage of all possible sun facing orientations.

In order to proceed with such a design, we will need to compute/define initial models of how we would combine the initial sensor inputs into comprehensible information that would provide the final orientations of our cube satellite.

Some assumptions that we are making for these initial designs are the following:
- Act as if our cube satellite is a perfect cube, with photodiodes at the center of each face
- Our sensors are perfect, so we aren't taking into any account for noise/error
- use intensity values as vectors moving out from the origin
